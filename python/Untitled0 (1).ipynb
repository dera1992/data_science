```python
import pandas as pd
DF_train=pd.read_csv('/content/sample_data/california_housing_train.csv')
DF_train.head()
```

```python
DF_train_final=DF_train.loc[:,['housing_median_age','total_rooms','total_bedrooms','population','households','median_income','median_house_value']]
DF_train_final
```

```python
x_train=DF_train_final.loc[:,['housing_median_age','total_rooms','total_bedrooms','population','households','median_income']]
x_train
```

```python
y_train=DF_train_final.loc[:,['median_house_value']]
y_train
```

```python
import pandas as pd
DF_test=pd.read_csv('/content/sample_data/california_housing_test.csv')
DF_test
```

```python
DF_test_final=DF_test.loc[:,['housing_median_age','total_rooms','total_bedrooms','population','households','median_income','median_house_value']]
DF_test_final
```

```python
x_test=DF_test_final.loc[:,['housing_median_age','total_rooms','total_bedrooms','population','households','median_income']]
x_test
```

```python
y_test=DF_test_final.loc[:,['median_house_value']]
y_test
```

```python
from sklearn.linear_model import LinearRegression
model_LR = LinearRegression()
model_LR.fit(x_train, y_train)
```

```python
y_predict = model_LR.predict(x_test)
y_predict
```

```python
from sklearn import metrics
import numpy as np
print('Mean Squared Error:',metrics.mean_squared_error(y_test,y_predict))
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_predict))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
```
